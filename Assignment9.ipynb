```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
penguins.corr()
```

```python
penguins.plot()
```

```python
penguins.corr().style.background_gradient(cmap='coolwarm', axis=None).set_precision(2)
```

```python
# Strongest positive correlation (excluding self)
penguins.plot.scatter(x='body_mass_g', y='flipper_length_mm')
```

```python
# Strongest negative correlation
penguins.plot.scatter(x='flipper_length_mm', y='bill_depth_mm')
```

```python
# Weakest correlation
penguins.plot.scatter(x='bill_depth_mm', y='bill_length_mm')
```
