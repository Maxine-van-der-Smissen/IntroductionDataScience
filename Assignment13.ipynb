```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
```

```python
from scipy.stats import chi2_contingency
```

```python
contingencyTableSpeciesSex = penguins.groupby(['species', 'sex']).size().unstack('sex', fill_value=0)
contingencyTableSpeciesSex
```

```python
contingencyTableSpeciesSex.plot.bar()
```

There seems to be a correlation between the number of male and female individuals, when looking at a certain species.

```python
chi2_contingency(contingencyTableSpeciesSex)
```

The null hypothesis states that both variables are independent of one another, whereas the alternative hypothesis states that they are dependent.
The p-value is 97.6%, this is a lot higher than a significance of 0.05 or 5%.
The null hypothesis can not be discarded, these variables are independent of each other.

```python
contingencyTableIslandSex = penguins.groupby(['island', 'sex']).size().unstack('sex', fill_value=0)
contingencyTableIslandSex
```

```python
contingencyTableIslandSex.plot.bar()
```
